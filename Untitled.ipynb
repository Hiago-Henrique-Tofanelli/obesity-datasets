```python
import numpy as np
```

```python
# Dados
states = obesityDT['NAME']
obesity_values = obesityDT['Obesity']

# Número de estados
num_states = len(states)

# Configurar o gráfico
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})

# Ângulo do setor do gráfico
angles = np.linspace(0, 2 * np.pi, num_states, endpoint=False).tolist()
angles += angles[:1]

# Plote os valores de obesidade
ax.fill(angles, obesity_values.tolist() + obesity_values.tolist()[:1], 'b', alpha=0.1)
ax.set_yticklabels([])

# Adicione rótulos para os estados
ax.set_xticks(angles[:-1])
ax.set_xticklabels(states, fontsize=10, ha='center')

# Adicione uma grade radial
ax.set_yticks([])

# Defina o título
plt.title('Comparação de Índice de Obesidade por Estado', size=16)

# Ajuste o layout para melhorar a legibilidade
plt.tight_layout()

# Mostre o gráfico
plt.show()
```

```python
# Dados
states = obesityDT['NAME']
obesity_values = obesityDT['Obesity']

# Gerar cores em degradê
n_states = len(states)
colors = plt.cm.get_cmap('coolwarm', n_states)(np.arange(n_states))

# Configurações do gráfico
fig, ax = plt.subplots(figsize=(13, 15))
ax.axis('equal')  # Proporção igual para que o gráfico seja um círculo

# Gráfico de pizza com cores em degradê
wedges, _ = ax.pie(
    obesity_values, labels=None, startangle=140, wedgeprops={'linewidth': 0}, colors=colors
)

# Adicionando os nomes dos estados em direção ao centro
for i, state in enumerate(states):
    angle = (wedges[i].theta2 - wedges[i].theta1) / 2.0 + wedges[i].theta1
    x = 0.7 * np.cos(np.deg2rad(angle))
    y = 0.7 * np.sin(np.deg2rad(angle))
    rotation = angle  # Ângulo de rotação igual ao ângulo de posição
    if angle > 90 and angle < 270:  # Evitar que os nomes fiquem de cabeça para baixo
        rotation += 180
    ax.annotate(state, xy=(x, y), fontsize=12, ha='center', va='center', color='black', rotation=rotation)  # Adicionando o nome

# Adicionando percentagens dentro do círculo
total_obesity = sum(obesity_values)
percentage_values = [f'{(value/total_obesity*100):.1f}%' for value in obesity_values]

for i, p in enumerate(percentage_values):
    angle = (wedges[i].theta2 - wedges[i].theta1) / 2.0 + wedges[i].theta1
    x = 0.5 * np.cos(np.deg2rad(angle))
    y = 0.5 * np.sin(np.deg2rad(angle))
    ax.annotate(p, xy=(x, y), fontsize=10, ha='center', va='center', color='black')  # Adicionando o percentual dentro do círculo

# Título do gráfico
plt.title('Índice de Obesidade por Estado', size=16)

# Exibição do gráfico
plt.show()
```
